```python
# all 1s matrix
np.ones((4,2,2))
```

```python
# any other number
import numpy as np
a = np.full((2,2), 99)
a
```

```python
#random decimal numbers 
np.random.rand(4,2)
```

```python
#random integers values
np.random.randint(-4,7, size=(3,3))
```

```python
# the identity matrix
np.identity(5)
```

```python
# repeat an array
arr = np.array([1,2,3])
r1 = np.repeat(arr,3, axis = 0)
print(r1)
```

```python
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
z[1,1] = 9
print(z)

output[1:4,1:4] = z
print(output)
```
